```python
# So Cal Yoga Emporium_Caroline Olearnek_5/24/2024
# Simulate 52 weeks of sales for the corporation and save to a CSV file
# Then drate a data frame that includes the full years sales and analyses and graphs results

# General Setup
import random
import pandas as pd
import os
from datetime import datetime, timedelta
import altair as alt


# Product Lists
items = ['Mat', 'Block', 'Blanket', 'Bolster', 'Strap', 'Towel', 'Water Bottle']
sizes = ['X-small','Small','Medium','Large','X-Large']
styles = ['Thick', 'Thin', 'Organic', 'Rugged']
colors = ['black', 'red', 'green', 'natural']
customer_names = ['Lilian', 'Caelen', 'George', 'Susan', 'Brent', 'Mia', 'Larry', 'David', 'Margret', 'Neldon', 'Ken', 'Paul', 'Christian', 'Catherine', 'Ronnie', 'Karen', 'Marion', 'Ernest', 'Lorna', 'Steven']
store_names = ['HB_YogaStore', 'Newport_YogaStore', 'Yorba_YogaStore', 'CostaMesa_YogaStore']

#Set stores
locations = 2

#classes

# Define Product Class
# The section sets up corporation products and prices
class YogaTool:
    def __init__(self):
        self.name = random.choice(items) + " " + random.choice(sizes) + " " + random.choice(styles) + " " + random.choice(colors)
        self.price = round(random.uniform(1,50),2)

# String representation of YogaTool
    def __str__(self):
        return self.name + " ($" + str(self.price) + ")"

# Define the Inventory class
# Create a random list of store products
class Inventory:
    def __init__(self):
        self.tools = []

        #Build the inventory
        num_tools = random.randint(10, 20) # Generate a random number of products between 10 and 20
        for i in range(num_tools):
            self.tools.append(YogaTool()) #Build product and add to inventory

# Display all tools in the inventory
    def display_inventory(self):
        for tool in self.tools:
            print(tool)

# Store Class
# This class makes a store that sells yoga tools, assigns inventory and lists customers
# This simulates customers shopping and making purchases
class Store:
    def __init__(self,name):
        self.name = name                      #add a name
        self.local_inventory = Inventory()    #add inventory
        self.customers = []                   #add customer
        self.purchase = Purchase()            #customer make purchase

    # Add a customer to the store
    def add_customer(self, customer):
        self.customers.append(customer)

    def getCustomers(self):
        for customer in self.customers:
            print("customer " + customer.getName() + " is busy shopping.")

    def run(self): # Run the simulation for the store
        print("\n" + "Welcome to " + self.name + "!\n")
        print(self.local_inventory.display_inventory())
        print("We have " + str(len(self.customers)) + " customers today.")

        for customer in self.customers: # Iterate over the customers
            print("Customer " + customer.name + " is shopping.")
            # Take one item from the inventory and add it to the customer's basket
            item = random.choice(self.local_inventory.tools)
            customer.add_to_cart(item)
            print("Customer " + customer.name + " added " + str(item) + " to their order.")
            #New Customer purchase
            self.purchase.increment(item.price) # Increment the sale object with the price of the item
            print("Customer " + customer.name + " checked out and paid $" + str(item.price) + ".")

        print("The total sales for " + self.name + " today is $" + str(self.purchase.total) + ".")

    #return store name and item
    def __str__(self):
        return self.name + " (Sales: $" + str(self.purchase.total) + ")"

    def show_store(self):
        print(self.local_inventory.display_inventory())


# Customer Class
# This class creates the customer
class Customer:
    def __init__(self,name):  # Initialize a customer with a name and a cart
        self.name = name
        self.cart = Cart()

    def getName(self):
        return self.name

    def add_to_cart(self, item): # Add a product to the customer's cart
        self.cart.add(item)

    # Return the name and the cart of the customer
    def __str__(self):
        return self.name + " (" + str(self.order) + ")"

# Cart Class
# This class creates a shopping basket and add items to the basket
class Cart: #slight variation on the store with cart

    def __init__(self): # Initialize a cart with a list of products
        self.products = []

    def add(self, item): # Add a product to the order
        self.products.append(item)

    def __str__(self): # Return the list of products in the order
        result = "Order:\n"
        for item in self.products:
            result += str(item) + "\n"
        return result

# Purchase Class
# This class creates sales for the store
class Purchase:
    # Initialize a sale with a total amount
    def __init__(self):
        self.total = 0

    # Increment the total amount by a given price
    def increment(self, price):
        self.total += price

    # Return the total amount of the sale
    def __str__(self):
        return "Sale: $" + str(self.total)

# Corporation Class
# This class defines the corporation and simulates multiple locations
# This class also generates weekly store sales data and writes it to a CSV file
class Corporation:
    # Initialize a corporation with a name and a list of stores
    def __init__(self, name):
        self.name = name
        self.stores = []
        self.analytics = []

    # Run the simulation for the corporation
    def simulate(self, week_name):
        print("Welcome to " + self.name + "!")
        print("We have " + str(locations) + " stores in our corporation.")
        # Iterate over the stores
        total_sales = 0
        for l in range(locations):
            name = store_names[l]
            location = Store(name)

            #store customer list
            num_cust = random.randint(1, 20)

            for i in range(num_cust):
                cust_name = random.choice(customer_names)
                location.add_customer(Customer(cust_name))

            location.run()
            self.stores.append(str(location))
            total_sales += (location.purchase.total)

            store_dict = {'store_name': location.name, 'weekly_sales_total': location.purchase.total}
            weekly_totals_per_store.append(store_dict)

            # Create Pandas DataFrame
            df = pd.DataFrame(weekly_totals_per_store)

            # Write DataFrame to a CSV file
            save_as = week_name.replace(' ', '-').lower() + '.csv'
            df.to_csv(save_as, index=False)

            print(f"Data saved to {save_as}")

            print("\n\n")

        # Display the total sales for the corporation
        print("The total sales for " + self.name + " today is $" + str(total_sales) + ".")


# Global Data Frame Class
# This class to create Global Data Frame for analytics and uses generated CSV files
# to compile weekly sales data for annual analysis and store it in memory.

class Create_Global_Dataframe:
    def __init__(self):
        # Get the current directory (where the Colab notebook is running)
        current_directory = os.getcwd()

        # List all files
        all_files = os.listdir(current_directory)

        # Filter only CSV files
        csv_files = [file for file in all_files if file.endswith('.csv')]

        # Initialize the master list
        all_stores_all_weekly_sales = []  #global dataframe

        # Read each CSV file, extract rows, and append to the master list
        for filename in csv_files:
            df = pd.read_csv(filename)
            week_name = filename.replace('.csv', '')  # Extract week name from filename
            for _, row in df.iterrows():
                row_dict = row.to_dict()
                row_dict['week'] = week_name
                all_stores_all_weekly_sales.append(row_dict)

        # Create a DataFrame from the list
        self.df = pd.DataFrame(all_stores_all_weekly_sales)

    def get_df(self):
        return self.df

# Sales Analysis Class
# This class analyses store sales data
class Sales_Analytics:
    def __init__(self):

        # Group the DataFrame by 'store_name' and calculate statistical properties
        grouped_stats = df.groupby('store_name')['weekly_sales_total'].describe()

        # Calculate the variance for each store
        grouped_variance = df.groupby('store_name')['weekly_sales_total'].var().rename("variance")

        # Combine the statistics and variance into one DataFrame
        self.stats_with_variance = pd.concat([grouped_stats, grouped_variance], axis=1)

        # Print the combined statistics and variance
        print(self.stats_with_variance)
        print("\n\n")

        # Calculate the variance for each store
        grouped_variance = df.groupby('store_name')['weekly_sales_total'].var().rename("variance")

        # Rank the stores by variance
        self.ranked_stores = grouped_variance.sort_values(ascending=False)

        # Print the ranked stores
        print("Stores Ranked By Largest Variance:")
        print(self.ranked_stores)

    def get_ranked_stores(self):
        return self.ranked_stores

# Main Code
# Name the Corporation
Corp = Corporation("So Cal Yoga Emporium")

# Generate weekly sales for every week in 2024
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)
current_date = start_date

while current_date <= end_date:
    weekly_totals_per_store = []
    week_name = current_date.strftime('%B-%d-%Y')
    current_date += timedelta(days=7)   # Move to the next week
    # print(current_date)               # this should be the name of your csv
    Corp.simulate(week_name)         # Execute Corporate Sim for the week

cgd = Create_Global_Dataframe()
df = cgd.get_df()

# Create the Yearly Sales Chart
# Define the width and height
width = 800  # Set the width to be twice of a typical value (e.g., 400)
height = 300  # Keep the height unchanged

# Create a base line chart with specified width and height
base = alt.Chart(df, width=width, height=height).mark_line().encode(
    x='week:T',
    y='weekly_sales_total:Q',
    color='store_name:N',
    tooltip=['store_name:N', 'weekly_sales_total:Q', 'week:T']
).interactive()


# Display the chart
base


```

```python
# *** Altair will only display charts if they are the last element for notebook environment
# *** To see both charts, cut and paste the code below into new window and execute

# Analyze Sales
sa = Sales_Analytics()
ranked_stores = sa.get_ranked_stores()

# Chart the Variance of eeach store
# Convert the ranked_stores Series to a DataFrame for Altair
data = ranked_stores.reset_index()
data.columns = ['store_name', 'variance']

# Define the color scale with a midpoint
color_scale = alt.Scale(scheme='redblue', domainMid=0)

# Create the histogram
histogram = alt.Chart(data).mark_bar().encode(
    x='store_name:N',
    y='variance:Q',
    color=alt.Color('variance:Q', scale=color_scale, legend=None),
    tooltip=['store_name:N', 'variance:Q']
).properties(
    width=800
)

# Display the histogram
histogram
```
